```python

class Person(object):
    def __new__(cls,*args,**kwargs):
        print('__new__clsid{0}'.format(id(cls)))
        obj = super().__new__(cls)
        print('id{0}'.format(id(obj)))
        return obj
    def __init__(self,name,age):
        print('__init__selfid{0}'.format(id(self)))
        self.name = name
        self.age = age

print('objectid{0}'.format(id(object)))
print('Personid{0}'.format(id(Person)))

#Person
p1 = Person('',20)
print('p1Personid{0}'.format(id(p1)))
```
