```python
#打印指定目录下的所有的.ipynb文件
path = os.getcwd()
lst = os.listdir(path)
for filename in lst:
    if filename.endswith('.ipynb'):
        print(filename)
                         
```

```python
#使用walk打印整个目录下的所有的文件
#遍历该路径下的所有目录和子文件，并且按顺序对应
path = os.getcwd()
print(path,'\n')
lst = os.walk(path)
for dirpath,dirname,filename in lst:
    print(dirpath) # 目录
    print(dirname) # 文件夹
    print(filename) # 文件
    print('-------------')
```

```python
#两个for，第一个是获取根目录所有的文件，
#第二个是获取根目录文件夹内的其他文件
path = os.getcwd()
print(path)
lst = os.walk(path)
for dirpath,dirname,filename in lst:
    for dir in dirname:
        print(os.path.join(dirpath,dir))
    #print('-------------')

    for file in filename:
        print(os.path.join(dirpath,file))
    print('-------------')
```
